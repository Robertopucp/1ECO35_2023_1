```python
## PARTE 3 


#función que permita calcular el factorial de un número (n)

def factorial(n):
    if n == 0:
        return 1
    else:
        resultado = 1
        for i in range(1, n + 1):
            resultado *= i
        return resultado

n = 5  # número hasta el que se desea generar la secuencia de factoriales
for i in range(1, n + 1):
    print(f"{i}! = {factorial(i)}")
```

```python
#Función de masa corporal

def calcular_imc(peso, talla):
    imc = peso / (talla ** 2)
    print("peso : {} kg".format(peso))
    print("talla : {} m".format(talla))
    print("IMC es: {:.2f}".format(imc))
    if 18.5 < imc < 24.9:
        print("Clasificación: Normal")
    elif 25 < imc < 29.9:
        print("Clasificación: Sobrepeso")
    elif 30 < imc < 34.9:
        print("Clasificación: Obesidad grado I")
    elif 35 < imc < 39.9:
        print("Clasificación: Obesidad grado II")
    elif 40 < imc: 
        print("Clasificación: Obesidad grado III")

# Ejemplo de uso de la función
calcular_imc(70, 1.75)  # Peso: 70 kg, Talla: 1.75 m
```

```python
##Alcanzamos al portfolio en la carpeta de data a través del ../ y de colocar las carpetas correspondientes
portfolio = pd.read_csv(r'../../data/Portafolio.csv')
portfolio
```

```python
def calculadora(data):
    X = data['X']
    Y = data['Y']
    w1 = 0.2
    w2 = 0.8
    pearson = np.corrcoef(X, Y)[0,1]
    varianza = (np.var(X)*w1**2) + (np.var(Y)*w2**2) + (2*np.cov(X,Y)*w1*w2)
    
    return [pearson, varianza]
```

```python
##Obtenemos el Pearson y la varianza
resultados = calculadora(portfolio)
print("Coeficiente de correlación de Pearson:", resultados[0])
print("Varianza:", resultados[1])
```

```python
#Keywords especiales _(*args and **kwargs)

def procesar_datos(*args, escalar=False, inversa_hiperbolica=False):
    datos = np.array(args)
    
    if escalar:
        datos = (datos - datos.mean()) / datos.std()
    elif inversa_hiperbolica:
        datos = np.arcsinh(datos)
        
    return datos
```
