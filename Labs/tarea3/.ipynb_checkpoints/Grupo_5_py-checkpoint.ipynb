Aplique la siguiente función a un vector de 100 numeros aleatorios y a las columnas numericas de la base de datos BDD_compras_consumidores.xlsx, donde min(x) y max(x) son el minimo y maximo valor del vector asi como de cada columna según corresponda. Por su parte, Xi representa cada número.

Primera parte, en el que trabajamos con el vector de 100 numeros aleatorios

```python
vector = np.arange(100)
def escalonamiento(x1,min,max):
    
    escalonamiento = (x1-min)/(max-min)
    
    return escalonamiento

list( map( lambda x1, min = np.min(vector), max = np.max(vector): escalonamiento(x1,min,max) , vector)  )  
```

a) Función para calcular el factorial de un número n!

```python
def factorial(n):
    if n < 0:
        return None # Devuelve None si el número es negativo
    elif n == 0:
        return 1
    else:
        fact = 1
        for i in range(1, n+1):
            fact *= i
        return fact
```

Probaremos con 4!

```python
factorial (4)
```

c) Función aplicado a dos activos financieros

- Coeficiente de pearson del rendimiento de los activos X e Y.

```python
import pandas as pd

# URL del archivo csv
url = "https://github.com/Robertopucp/1ECO35_2023_1/raw/main/data/Portafolio.csv"

# Cargar el archivo csv desde la URL
portfolio = pd.read_csv(url)
portfolio

X = portfolio['X']
y = portfolio['Y']

import numpy as np

def alpha(data):
    X = data['X']
    y = data['Y']
    cov = np.cov(X, y)[0, 1]
    var_X = np.var(X, ddof=1)
    var_y = np.var(y, ddof=1)
    
    coeficiente = cov / np.sqrt(var_X) * np.sqrt(var_y)
    return coeficiente

alpha(portfolio)
```

- Varianza del portafolio

```python
wx = 0.2
wy = 0.8

def alpha2(data):
    X = data['X']
    y = data['Y']
    var_X = np.var(X, ddof=1)
    var_y = np.var(y, ddof=1)
    cov = np.cov(X, y)[0, 1]
    varianza = wx**2*var_X + wy**2*var_y + 2*wx*wy*cov
    return varianza

alpha2(portfolio)
```

d) Construya una función que permite una cantidad no restringida de datos (*args) y que admita dos tipos de métodos que puedan aplicarse a la cantidad no restringida de datos (* *kargs). Los métodos que debe admintir la función son el escalamiento de la primera pregunta y la transformación inverse hyperbolic sine:

```python
import numpy as np

def apply_methods(*args, **kwargs):
    """
    Esta función aplica los métodos de escalamiento y la transformación inverse hyperbolic sine a los datos proporcionados.
    :param args: datos a los que se aplicarán los métodos
    :param kwargs: argumentos con los métodos a aplicar
    :return: los datos transformados
    """
    if not args:
        raise ValueError("Se requiere al menos un argumento")

    # Se comprueba si se proporciona el método de escalamiento (también llamado fórmula de normalización min-max)
    if 'min_max' in kwargs and kwargs['min_max'] is True:
        # Se aplica la fórmula de escalamiento
        data_norm = []
        for data in args:
            data_norm.append((data - np.min(data)) / (np.max(data) - np.min(data)))
        args = data_norm

    # Se comprueba si se proporciona el método de transformación inverse hyperbolic sine
    if 'asinh' in kwargs and kwargs['asinh'] is True:
        # Se aplica la transformación inverse hyperbolic sine:
        data_asinh = []
        for data in args:
            data_asinh.append(np.log(data + np.sqrt(data**2 + 1)))
        args = data_asinh

    return args
```

Si se proporciona el argumento min_max=True, la función aplica el escalamiento de la primera pregunta a los datos proporcionados. Si se proporciona el argumento asinh=True, la función aplica la inverse hyperbolic sine a los datos proporcionados.
