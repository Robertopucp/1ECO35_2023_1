```python
import pandas as pd 
import numpy as np 
import os
import pyreadstat
os.getcwd() 
data, meta = pyreadstat.read_sav(r'../../data/actos_est.sav', metadataonly=False)
df = pd.DataFrame(data) 
```

```python
#Mostrar las etiquetas de valor y las etiquetas de las variables
print(meta.value_labels)
print(meta.variable_value_labels)
```

```python
#Filtrar la base de datos según los siguientes tipos de eventos: desaparición, secuestros, reclutamiento forzado y muertes en atentados.
filtros = data[(data['IDTIPOAC'] == 'LDS') | (data['IDTIPOAC'] == 'LSE') | 
                (data['IDTIPOAC'] == 'LRC') | (data['IDTIPOAC'] == 'MAT')]
filtros
```

```python
#Borrar los duplicados de la columna (IDACTO). 
data = data.drop_duplicates(subset=['IDACTO'])
print(data)
```

```python
#Crear una base de datos con el total de eventos de violencia según su tipo a nivel departamento (DEPNAO).
base_datos_DEPNAO = data.groupby('DEPNAO').size().reset_index(name='Total')
print(base_datos_DEPNAO)
```

```python
#Crear una base de datos con el total de eventos de violencia según su tipo a nivel distrito (UBIDIST).
base_datos_UBIDIST = data.groupby('UBIDIST').size().reset_index(name='Total')
print(base_datos_UBIDIST)
```

```python
#Crear una base de datos con el total de eventos de violencia según su tipo a nivel departamento (DEPNAO) y periodo (PERIODO). 
base_datos_DEPNAO_PERIODO = data.groupby(['DEPNAO', 'PERIODO']).size().reset_index(name='Total')
print(base_datos_DEPNAO_PERIODO)
```
