```python
#Procesamos la base de datos de la Comisión de la Verdad y Reconciliación (CVR). 
```

```python
#Importamos las librerías
import pandas as pd
import numpy as np
import os
import pyreadstat
```

```python
###en caso de no tener el pyreadstat, instalarlo con pip install pyreadstat
```

```python
#Luego cargamos la base de datos 
cvr, meta =  pyreadstat.read_sav( r"../../data/actos_est.sav")
cvr
```

```python
cvr.info()
```

```python
## 1) Etiquetas de valor y etiquetas de las variables
```

```python
# Extraemos las etiquetas de variable en var_labels
var_labels=dict(zip(meta.column_names, meta.column_labels))
var_labels
```

```python
# Extaemos las etiquetas de valor
var_values=meta.variable_value_labels
var_values
```

```python
## 2) Filtrar la base de datos 
```

```python
#Desaparición
cvr[cvr.IDTIPOAC=="LDS"]
```

```python
#Secuestros
cvr[cvr.IDTIPOAC=="LSE"]
```

```python
#Reclutamiento forzado
cvr[cvr.IDTIPOAC=="LRC"]
```

```python
#Muertes en atentados
cvr[cvr.IDTIPOAC=="MAT"]
```

```python
## 3) Borrar duplicados
```

```python
#Solo dejamos la primera observacion de cada persona
cvr2 = cvr.drop_duplicates(subset=["IDACTO"], keep='first')
cvr2
```

```python
##4) Crear distintas base de datos
###4.1) Total de eventos de violencia según su tipo a nivel departamento
```

```python
# Buscamos qué tipo de actos se registran en la base de datos
actos=var_values['IDTIPOAC']
actos
```

```python
# Después, buscamos los departamentos registrados
var_values['DEPNA0']
```

```python
#La base de datos tiene las dummys necesarias, entonces  solo agregamos las observaciones
dep=cvr2.groupby(['DEPNA0']).agg(LDS=('LDS_LDT', np.sum), LDT = ('LDT', np.sum), LRC = ('LRC', np.sum), LSE = ('LSE', np.sum), MAE = ('MAE', np.sum), MAT = ('MAT', np.sum), MEF = ('MEF', np.sum), TLA = ('TLA', np.sum), TTR = ('TTR', np.sum), TVS = ('TVS', np.sum))
dep
```

```python
#Editamos la DF anterior para ver los nombres de los departamentos
dep2=dep.rename(index={1.0: 'AYACUCHO',
 2.0: 'APURÍMAC',
 3.0: 'HUANCAVELICA',
 4.0: 'CUSCO',
 5.0: 'HUÁNUCO',
 6.0: 'UCAYALI',
 7.0: 'SAN MARTÍN',
 8.0: 'PUNO',
 9.0: 'JUNÍN',
 10.0: 'LIMA - CALLAO',
 90.0: 'OTROS'})
dep2
```

```python
### 4.2) Total de eventos de violencia según su tipo a nivel distrito
```

```python
#Hacemos lo mismo que anteriormente
dist=cvr2.groupby(['UBIDIST']).agg(LDS=('LDS_LDT', np.sum), LDT = ('LDT', np.sum), LRC = ('LRC', np.sum), LSE = ('LSE', np.sum), MAE = ('MAE', np.sum), MAT = ('MAT', np.sum), MEF = ('MEF', np.sum), TLA = ('TLA', np.sum), TTR = ('TTR', np.sum), TVS = ('TVS', np.sum))
dist
```

```python
#Calculamos el logaritmo natural de las variables
dist_log=np.log(dist)
dist_log
```
