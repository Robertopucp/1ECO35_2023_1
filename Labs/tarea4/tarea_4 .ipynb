```python
# se filtra la base de datos para los tipos de eventos: desaparición, secuestros, reclutamiento forzado y muertes en atentados
datospss, meta =  pyreadstat.read_sav(r"C:\Users\nmiyamoto\Documents\GitHub\1ECO35_2023_1\data\actos_est.sav")
cvr_datos = datospss[(datospss.IDTIPOAC == "LDS_LDT")| (datospss.IDTIPOAC == "LSE")| (datospss.IDTIPOAC == "LRC")| (datospss.IDTIPOAC == "MAT")]
cvr_datos
```

```python
# se eliminan los valores duplicados de la columna IDACTO
cvr_datos = cvr_datos.copy() 
# se colocó el .copy para no recibir advertencia del SettingWithCopyWarning
cvr_datos.drop_duplicates(subset='IDACTO', inplace=True)
cvr_datos.reset_index(drop=True, inplace=True)
cvr_datos
```

```python
# se crea una nueva base de datos filtrando idacto, idtipoac y departamento
df_nuevo = cvr_datos[['IDACTO','IDTIPOAC', 'DEPNA0','UBIDIST','PERIODO','LDS_LDT', 'LSE', 'LRC', 'MAT']]

# se renombra el depnao
df_nuevo = df_nuevo.rename(columns={'DEPNA0': 'DPTO', "UBIDIST": "DIST", "PERIODO": "PER"})
df_nuevo
```

```python
# para cada departamento se obtiene el total de eventos de violencia

tipo_event_dpto = pd.pivot_table(df_nuevo, index='DPTO', values=['LDS_LDT', 'LSE', 'LRC', 'MAT'],
                              aggfunc=np.sum, margins=True, margins_name='Total')
tipo_event_dpto

# se renombra cada evento
df_nuevorename = tipo_event_dpto.rename(columns={'LDS_LDT': 'TotalDesap','LSE': 'TotalSecu','LRC': 'TotalRe','MAT': 'TotalMuertes'})
df_nuevorename
```

```python
# para cada distrito se obtiene el total de eventos de violencia

tipo_event_dist = pd.pivot_table(df_nuevo, index='DIST', values=['LDS_LDT', 'LSE', 'LRC', 'MAT'],
                              aggfunc=np.sum, margins=True, margins_name='Total')
tipo_event_dist

# se renombra cada evento
df_nuevorename = tipo_event_dist.rename(columns={'LDS_LDT': 'TotalDesap','LSE': 'TotalSecu','LRC': 'TotalRe','MAT': 'TotalMuertes'})
df_nuevorename
```

```python
# crear variables con el logaritmo del total de eventos
df_nuevorename['Log_TotalDesap'] = np.log(df_nuevorename['TotalDesap'])
df_nuevorename['Log_TotalSecu'] = np.log(df_nuevorename['TotalSecu'])
df_nuevorename['Log_TotalRe'] = np.log(df_nuevorename['TotalRe'])
df_nuevorename['Log_TotalMuertes'] = np.log(df_nuevorename['TotalMuertes'])
df_nuevorename

# los valores que salen -inf significa que se intentó calcular logaritmos de valores de 0 por ello sale el warning
```

```python
# para cada departamento y periodo se obtiene el total de eventos de violencia

tipo_event_dptoyper = pd.pivot_table(df_nuevo, index= ['DPTO','PER'], values=['LDS_LDT', 'LSE', 'LRC', 'MAT'],
                              aggfunc=np.sum, margins=True, margins_name='Total')
tipo_event_dptoyper

# se renombra cada evento
df_nuevorename = tipo_event_dptoyper.rename(columns={'LDS_LDT': 'TotalDesap','LSE': 'TotalSecu','LRC': 'TotalRe','MAT': 'TotalMuertes'})
df_nuevorename
```
