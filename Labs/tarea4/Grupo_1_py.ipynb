En este script procesamos la base de datos de la Comisión de la Verdad y Reconciliación (CVR). Esta base de datos es el registra actos de violencia perpetrados contra individuos.

```python
#Primero importamos las librerías necesarias
import pandas as pd
import numpy as np
import os
import pyreadstat
```

```python
#Luego cargamos la base de datos 
cvr, meta =  pyreadstat.read_sav( r"../../data/actos_est.sav")
cvr
```

Se puede ver que tenemos 118 variables, con 36018 observaciones.

```python
cvr.info()
```

## 1. Mostar las etiquetas de valor y etiquetas de las variables

```python
# Extraemos las etiquetas de variable en var_labels
var_labels=dict(zip(meta.column_names, meta.column_labels))
var_labels
```

```python
# Del mismo modo, extaemos las etiquetas de valor
var_values=meta.variable_value_labels
var_values
```

## 3. Filtrar la base de datos 
Filtramos la base de datos depentiendo del tipo de acto de violencia indicado.

```python
#Desaparición
cvr[cvr.IDTIPOAC=="LDS"]
```

```python
#Secuestos
cvr[cvr.IDTIPOAC=="LSE"]
```

```python
#Reclutamiento forzado
cvr[cvr.IDTIPOAC=="LRC"]
```

```python
#Muertes en atentados
cvr[cvr.IDTIPOAC=="MAT"]
```

## 3. Borrar los duplicados
Borramos los duplicados de la columna IDACTO (ID del evento). Esto a modo de trabajar con la información de personas que sufrieron actos de violencia, en lugar de la cantidad de actos registrados.

```python
#nos quedamos con la primera observacion de cada persona
cvr2 = cvr.drop_duplicates(subset=["IDACTO"], keep='first')
cvr2
```

Se puede ver que ahora el número de observaciones se ha reducido

## 4. Crear distintas base de datos
### 4.1. Total de eventos de violencia según su tipo a nivel departamento

```python
# Primero, buscamos qué tipo de actos se registran en la base de datos
actos=var_values['IDTIPOAC']
actos
```

```python
# Luego, buscamos los departamentos registrados
# notar que la variable DEPNA0 registra el lugar de nacimineto de la víctima
var_values['DEPNA0']
```

```python
#La base de datos ya contiene las dummys necesarias, así que solo agregamos las observaciones
dep=cvr2.groupby(['DEPNA0']).agg(LDS=('LDS_LDT', np.sum), LDT = ('LDT', np.sum), LRC = ('LRC', np.sum), LSE = ('LSE', np.sum), MAE = ('MAE', np.sum), MAT = ('MAT', np.sum), MEF = ('MEF', np.sum), TLA = ('TLA', np.sum), TTR = ('TTR', np.sum), TVS = ('TVS', np.sum))
dep
```

```python
#Editamos la DF anterior para poder ver los nombres de los departamentos
dep2=dep.rename(index={1.0: 'AYACUCHO',
 2.0: 'APURÍMAC',
 3.0: 'HUANCAVELICA',
 4.0: 'CUSCO',
 5.0: 'HUÁNUCO',
 6.0: 'UCAYALI',
 7.0: 'SAN MARTÍN',
 8.0: 'PUNO',
 9.0: 'JUNÍN',
 10.0: 'LIMA - CALLAO',
 90.0: 'OTROS'})
dep2
```

### 4.2. Total de eventos de violencia según su tipo a nivel distrito

```python
#Replicamos el proceso anterior
dist=cvr2.groupby(['UBIDIST']).agg(LDS=('LDS_LDT', np.sum), LDT = ('LDT', np.sum), LRC = ('LRC', np.sum), LSE = ('LSE', np.sum), MAE = ('MAE', np.sum), MAT = ('MAT', np.sum), MEF = ('MEF', np.sum), TLA = ('TLA', np.sum), TTR = ('TTR', np.sum), TVS = ('TVS', np.sum))
dist
```

```python
#Calculamos el logaritmo natural de las variables
dist_log=np.log(dist)
dist_log
```
