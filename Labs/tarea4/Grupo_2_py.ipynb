## <a id='1.4'> 1.4 Crear una base de datos con el total de eventos de violencia según su tipo a nivel departamento (DEPNAO) </a> 

```python
#Se importan las librerías necesarias
import pandas as pd

#Se crea un diccionario con etiquetas para cada departamento
departamentos = {
    "01": "Amazonas", "02": "Ancash", "03": "Apurimac", "04": "Arequipa", "05": "Ayacucho", "06": "Cajamarca",
    "07": "Callao", "08": "Cusco", "09": "Huancavelica", "10": "Huanuco", "11": "Ica", "12": "Junin",
    "13": "La Libertad", "14": "Lambayaque", "15": "Lima", "16": "Loreto", "17": "Madre de Dios", "18": "Moquegua",
    "19": "Pasco", "20": "Piura", "21": "Puno", "22": "San Martin", "23": "Tacna", "24": "Tumbes", "25": "Ucayali"
}

#Se crea la variable DEPNAO que indica el departamento de cada evento
datossav['DEPNAO'] = datossav['UBIDIST'].str.slice(stop=2).replace(departamentos)

#Se crea la base de datos con el total de de eventos de violencia por departamento
datosViolenciaDEP = datossav.groupby('DEPNAO').agg(
    desapForz = ('LDS_LDT', 'sum'),
    detencion = ('LDT', 'sum'),
    reclutamientoForz = ('LRC', 'sum'),
    secuestro = ('LSE', 'sum'),
    asesinato = ('MAE', 'sum'),
    muerteAtentado = ('MAT', 'sum'),
    muerteEnfrentamiento = ('MEF', 'sum'),
    lesiones = ('TLA', 'sum'),
    tortura = ('TTR', 'sum'),
    violacionSexual = ('TVS', 'sum')
).reset_index()
datosViolenciaDEP #Se presenta el nuevo dataframe
```

## <a id='1.5'> 1.5 Crear una base de datos con el total de eventos de violencia según su tipo a nivel distrito (UBIDIST) </a> 


```python
#Se importan las librerías necesarias
import pandas as pd
import numpy as np

#Se crea la base de datos con el total de de eventos de violencia por distrito
datosViolenciaDIS = datossav.groupby('UBIDIST').agg(
    desapForz = ('LDS_LDT', 'sum'),
    detencion = ('LDT', 'sum'),
    reclutamientoForz = ('LRC', 'sum'),
    secuestro = ('LSE', 'sum'),
    asesinato = ('MAE', 'sum'),
    muerteAtentado = ('MAT', 'sum'),
    muerteEnfrentamiento = ('MEF', 'sum'),
    lesiones = ('TLA', 'sum'),
    tortura = ('TTR', 'sum'),
    violacionSexual = ('TVS', 'sum')
).reset_index()

#Se crea nuevas variables que calculan el logaritmo de las anteriores y se añaden al dataframe. Se decide utilizar np.log1p(x) para evitar calcular logaritmo de 0 dado que son dummy
datosViolenciaDIS[['log_desapForz', 'log_detencion', 'log_reclutamientoForz', 'log_secuestro', 'log_asesinato', 'log_muerteAtentado', 'log_muerteEnfrentamiento', 'log_lesiones', 'log_tortura', 'log_violacionSexual']] = \
    datosViolenciaDIS[['desapForz', 'detencion', 'reclutamientoForz', 'secuestro', 'asesinato', 'muerteAtentado', 'muerteEnfrentamiento', 'lesiones', 'tortura', 'violacionSexual']].apply(lambda x: np.log1p(x))
datosViolenciaDIS #Se presenta el nuevo dataframe
```
