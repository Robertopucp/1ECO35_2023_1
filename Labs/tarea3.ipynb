```python
#Creamos una funcion que de la masa corporal dado unos inputs 
def masa_corporal(p,t,e):
  if 18.5 < p/t**2 < 24.9:
    c = "Normal"
  if 25 < p/t**2< 29.9:
    c = "Aumentado"
  if 30 < p/t**2 < 34.9:
    c = "Moderado"
  if 35 < p/t**2 < 39.9:
    c = "Severo"
  if 40 < p/t**2:
    c = "Muy Severo"
  print("El estudiante",e, "tiene un peso de ", p, "y una talla de", t, "por lo tanto su indice es de ",round(p/t**2,2), "y su clasificacion es",c)

b = masa_corporal(70,1.5,1)
b
```

```python
# Usamos la funcion facotrial 
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)

#Creamos Una funcion que calcule la hyperbolic
import numpy as np

def hyperbolic(a):
    return np.log(a + np.sqrt(a**2+1))

#Creamos una funcion que calule el factorial o la hyperbolic de una serie de datos indefinida 
def funcion(*args, **kwargs): #definimos la funcion 
    method = kwargs.get('method', 'factorial') # usamos a factorial como predeterminado
    resultado = [] # creamos un loop para todos los resultados y los guardamos 
    for i in args:
        if method == 'factorial':
            res = factorial(i)
        elif method == 'hyperbolic':
            res = hyperbolic(i)
        resultado.append(res)
    print(resultado)   

```

```python

funcion(2,5, method='factorial')
```
