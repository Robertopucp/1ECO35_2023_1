## <a id='1.1'> 1.1 Formar una base de datos a partir de las bases data.dta y unidos.dta. Para ello, debe transformar la base data.dta a un formato long, luego hacer merge con la base unidos.dta. En el proceso de formato wide a long, use la variable numper como identificador </a> 

### <a id='1.1.1'> 1.1.1 Se transforma la base data.dta a un formato long </a> 

```python
#Se lee la base de datos
panelData = pd.read_stata("../../data/Juntos_program/data.dta",
                      convert_categoricals=False)

```

```python
#Se le añade un _ a los numpanh para poder utilizar luego el separador _
new_column_names = [col.replace('numpanh', 'numpanh_') for col in panelData.columns]
panelData.columns = new_column_names
panelData
```

```python
#Se filtra para quitar el separador para luego poder crear la variable period 
filter_list = list( map( lambda x: re.sub('.{3}$',"", x) , list(panelData.columns)[1::] )   )

new_list = list(dict.fromkeys(filter_list))
new_list
```

```python
#Se convierte de formato wide a long la base de datos data.dta
reshape_panelData = pd.wide_to_long(panelData, stubnames = new_list, i = ['numper'] , 
                                j = 'period' , sep = '_').reset_index()
#####
reshape_panelData
#Se ordena el dataframe
reshape_panelData = reshape_panelData.sort_values(['numper', 'period'], ascending=[True, True])
#Se resetea el index
reshape_panelData = reshape_panelData.reset_index(drop=True)


reshape_panelData
```

### <a id='1.1.2'> 1.1.2 Se realiza el merge con unidos.dta </a> 

```python
#Se lee la base de datos
baseUnidos = pd.read_stata("../../data/Juntos_program/unidos.dta",
                      convert_categoricals=False)
 
baseUnidos
```

```python
#Se realiza el merge de ambas bases de datos
baseDatos=pd.merge(reshape_panelData, baseUnidos, how = "left", on = "ubigeo", validate = "m:1")
baseDatos

```
