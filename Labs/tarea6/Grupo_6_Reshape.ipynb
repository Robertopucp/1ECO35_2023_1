## <a id='1.0'> 2. Reshape </a> 

```python
#definimos un diccionario llamado "new_columns" que mapea los nombres de las columnas antiguas a los nuevos nombres de columna 
new_columns = {'numpanh15': 'nump_15', 'numpanh16': 'nump_16', 'numpanh17': 'nump_17',
               'numpanh18': 'nump_18', 'numpanh19': 'nump_19'}
#Pasamos el diccionario "new_columns" como argumento para realizar la operación de renombrado.
juntos = juntos .rename(columns=new_columns)
columnas3 = juntos .columns.values
columnas3
```

```python
#transformamos la base data.dta "juntos" a un formato long
reshape_juntos = pd.wide_to_long(juntos, ['nump', 'mes', 'ubigeo', 'dominio', 'p400a3', 'p4022'],
                              i='numper', j='period', suffix='.+', sep='_').reset_index()
reshape_juntos
```

```python
#Calculamos el valor de 'year' sumando 2000 al valor de la columna 'period'
reshape_juntos['year'] = 2000 + reshape_juntos['period']
reshape_juntos
```

```python
# utilizamos el método size() para obtener el recuento de cada valor único en la columna 'year'.
num_year = reshape_juntos.groupby('year').size()
#imprimimos year_counts para mostrar el recuento de cada año en la columna 'year'
print(num_year )
```

## <a id='1.0'> 3. Merge </a> 

```python
#utilizamos el método merge() para realizar una combinación basada de la nueva base de datos "reshape_juntos" y "juntos2 
#para lograr ello utilizamos la columna común 'ubigeo'
base = reshape_juntos.merge(juntos2, on='ubigeo')
base
```

```python
columnas4 = base.columns.values
columnas4
```

```python
#creamos una lista llamada columnas_oficial que contiene los nombres de las columnas que deseamos mantener en el DataFrame "base"
columnas_oficial = ['year', 'numper', 'nump', 'mes', 'ubigeo', 'distrito', 'provincia', 'region', 'unidos',
                          'dominio', 'p400a3', 'p4022']
base = base[columnas_oficial]
base
```
