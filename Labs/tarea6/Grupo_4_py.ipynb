## 1. Reshape - Python

### Objetivo: Formar una base de datos a partir de las bases data.dta y unidos.dta.

```python
# 1. Importar la base de la ENAHO (panel 2015-2019) (base formato wide)
panel = pd.read_stata("../../data/Juntos_program/data.dta",
                      convert_categoricals=False)
```

```python
panel
```

```python
# Luego, nos quedamos solo con algunas variables
panel.columns = map(str.lower, panel.columns)  
print( panel.columns)
```

```python
# Se crea un identificador del hogar 
panel['numper'] = np.arange(1, panel.shape[0] + 1)
```

```python
panel
```

```python
# new_list: debe contener las variables en comun en todos los años (sin _año)
panel.columns[1::]
```

```python
filter_list = list( map( lambda x: re.sub('.{3}$',"", x) , list(panel.columns)[1::] )   )

# Se procede a eliminar los duplicados en la lista
new_list = list(dict.fromkeys(filter_list))

# esto último elimina duplicados
```

```python
filter_list
```

```python
new_list
```

```python
# 2. Reshape Wide to long 

reshape_panel = pd.wide_to_long(panel, stubnames = new_list, i = ['numper'] , 
                                j = 'period' , sep = '_').reset_index()
```

```python
# pasamos un formato WIDE a uno long
reshape_panel
```

```python
# Drop a las siguientes variables para que quede más limpio

del reshape_panel['period']
del reshape_panel['numpan']
```

```python
reshape_panel
```

```python
# Obtener una lista de todas las columnas en reshape_panel
columnas = reshape_panel.columns.tolist()

# Crear una nueva lista de columnas con las modificaciones deseadas
nuevas_columnas = []

for columna in columnas:
    if re.match(r'numpanh\d+', columna):
        nuevas_columnas.append('numpanh')
    else:
        nuevas_columnas.append(columna)

# Actualizar las columnas en reshape_panel
reshape_panel.columns = nuevas_columnas

# Eliminar las columnas duplicadas
reshape_panel = reshape_panel.loc[:, ~reshape_panel.columns.duplicated()]

reshape_panel
```

```python
# 3. Base de datos unidos.dta es una lista de todos los distritos del Perú

panel2 = pd.read_stata("../../data/Juntos_program/unidos.dta",
                      convert_categoricals=False)
```

```python
panel2
```

```python
# 4. Por último, realizamos el merge
merged_data = pd.merge(reshape_panel, panel2, on='ubigeo', how='inner')
```
