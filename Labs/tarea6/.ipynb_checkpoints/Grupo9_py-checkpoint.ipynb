```python
import pandas as pd
import numpy as np
import re 
import warnings
warnings.filterwarnings('ignore') 
```

```python
#Abrimos la Base de Datos 
panel1 = pd.read_stata("../../data/Juntos_program/data.dta",
                      convert_categoricals=False)
#2015-2019
```

```python
panel1
#Vemos que está en formato wide
#Está a nivel de individuo
```

```python
# nombre de las variables en minuscula

panel1.columns = map(str.lower, panel1.columns)  


print( panel1.columns)
```

```python
# Nos quedamos con todas las variables 

index_columns = np.where( panel1.columns.str.contains(
    '(numper)|(numpanh)|(mes_)|(ubigeo_)|(dominio_)|(p400a3_)|(p4022_)',
    regex=True))[0]

# [0] extraer la lista de posiciones dentro de la tupla con las psicicones de las columnas que satisfacen la condición 
```

```python
index_columns
```

```python
panel1 = panel1.iloc[: ,index_columns] 
# Renombramos las variables de año de nacimiento, si estuvo enfermo en los últimos 4 meses, el año de nacimiento y
#el identificador del hogar
    
panel1.rename(columns = {'p400a3_15':'añcmto_15', 'p400a3_16':'añcmto_16', 'p400a3_17':'añcmto_17',
                         'p400a3_18':'añcmto_18', 'p400a3_19':'añcmto_19', 'p4022_15':'enfermo_15',
                         'p4022_16':'enfermo_16','p4022_17':'enfermo_17','p4022_18':'enfermo_18',
                         'p4022_19':'enfermo_19', 'numpanh15': 'numhog_15','numpanh16': 'numhog_16',
                        'numpanh17': 'numhog_17','numpanh18': 'numhog_18','numpanh19': 'numhog_19',}, inplace = True)


```

```python
panel1
```

```python
cols = panel1.columns.tolist()
```

```python
#el identificador ya está en la primera columna
cols[0:-1]
```

```python
panel1.columns[1::]
```

```python
#filtramos y nos quedamos sin los últimos 3 caracteres _año
filter_list = list( map( lambda x: re.sub('.{3}$',"", x) , list(panel1.columns)[1::] )   )
new_list = list(dict.fromkeys(filter_list))
```

```python
new_list
```

```python
#Hacemos el reshape
reshape_panel = pd.wide_to_long(panel1, stubnames = new_list, i = ['numper'] , 
                                j = 'periodo' , sep = '_').reset_index()
```

```python
reshape_panel
```

```python
#Abrimos la otra Base de Datos 
panel2 = pd.read_stata("../../data/Juntos_program/unidos.dta",
                      convert_categoricals=False)
```

```python
panel2
```

```python
panel2.columns = map(str.lower, panel2.columns)  


print( panel2.columns)
```

```python
print(reshape_panel.columns)
print(panel2.columns)
```

```python
#Hacemos el merge con ubigeo
merged_panel = pd.merge(reshape_panel, panel2, on=['ubigeo'])
```

```python
merged_panel
```

```python
merged_panel['enfermo'] = merged_panel['enfermo'].fillna(-1)

# Convertimos "enfermo" a entero
merged_panel['enfermo'] = merged_panel['enfermo'].astype(int)
```
