En este script hacemos el reshape de la base de datos del programa juntos (de wiade a long) y el merge con la lista de distritos del Perú


```python
#Primero importamos todas las librerias necesarias
import pandas as pd
import numpy as np
import re 
import warnings
warnings.filterwarnings('ignore') 
```

### 1. Importamos las bases de datos

```python
#Importamos las bases de datos del programa juntos
data= pd.read_stata("../../data/Juntos_program/data.dta",
                      convert_categoricals=False)
data
```

```python
data.columns.values
```

```python
unidos= pd.read_stata("../../data/Juntos_program/unidos.dta",
                      convert_categoricals=False)
unidos
```

```python
unidos.columns.values
```

### 2. Reshape

```python
#Renombramos las varaibles necesarias 
data.rename(columns={'numpanh15': 'nump_15','numpanh16': 'nump_16',
                     'numpanh17': 'nump_17','numpanh18': 'nump_18',
                     'numpanh19': 'nump_19'}, inplace=True)
data.columns.values
```

```python
reshape_data = pd.wide_to_long(data, stubnames = ['nump','mes','ubigeo','dominio','p400a3','p4022'],
                                 i = ['numper'] , j = 'period' , sep = '_').reset_index()
```

```python
reshape_data
```

```python
reshape_data['year']=2000+reshape_data['period']
reshape_data
```

```python
print(reshape_data['year'].value_counts())
```

### 3. Merge

```python
#Para el merge se utilizará la variable ubigeo
base_final=pd.merge(reshape_data,unidos, on='ubigeo')
base_final
```

```python
#Reordeanmos la base
base_final.columns.values
```

```python
base_final=base_final[['year','numper','nump', 'mes', 'ubigeo','distrito', 'provincia', 'region', 'unidos','dominio', 'p400a3',
       'p4022']]
base_final
```
