```python
#Se importa las librerías necesarias
import camelot as cm  # libreria de camelot
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import os
from PyPDF2 import PdfFileReader
import re # expresiones regularesb
```

###Se extrae las tablas de G2_HUANUCO.pdf

```python
#Se extrae las tablas eliminando el caracter especial del espacio \n
tables = cm.read_pdf("/content/sample_data/G2_HUANUCO.pdf", pages = "1-8" , flavor = "lattice",
                     table_regions=['50,500,750,100'],  strip_text='\n')
```

```python
#Se prueba los resultados
tables[7].df
```

```python
#Se convierte la primera fila a columnas de variables
data = tables[0].df
data.columns = data.iloc[0]
data = data[1:]
data.columns = ['numero', 'anioJEC', 'codigoModular', 'codigoLocal', 'nombreIE', 'DRE', 'UGEL',
       'ubigeo', 'departamento', 'provincia', 'distrito', 'direccion','areaGeografica']
data
```

```python
#Se observa el tipo de tables
print(type(tables))
```

###Se vuelve todo un solo dataframe




```python
# Se crea una lista para poder almacenar las tablas convertidas en dataframes
dataframes = []

# Convierte todas las tablas en dataframes considerando la primera fila como las variables
for table in tables:
    df = table.df
    df.columns = df.iloc[0]
    df = df[1:]
    df.columns = ['numero', 'anioJEC', 'codigoModular', 'codigoLocal', 'nombreIE', 'DRE', 'UGEL',
       'ubigeo', 'departamento', 'provincia', 'distrito', 'direccion','areaGeografica']
    dataframes.append(df)

# Se combina todos los dataframes en uno solo llamado dataFrame
dataFrame = pd.concat(dataframes, ignore_index=True)

#Se elimina la variable numero innecesaria
dataFrame = dataFrame.drop(columns=["numero"])


# Se presenta el dataframe resultante
dataFrame

```

```python
#Se exporta en formato dta sin que se genere una columna de index en el dta
dataFrame.to_stata("jec_data_Grupo_2.dta",write_index=False)
```
