# Replicar el siguiente gráfico (producción de hoja de coca en hectáreas)

```python
# Leer el archivo Excel
prod_coca = pd.read_excel("../../data/produccion_coca/6.1.1_-_Illicit_coca_bush_cultivation.xlsx",
                          skiprows=lambda x: x not in [ 2, 4, 5, 6])
prod_coca
```

```python
variables = ['País', '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020']

prod_coca = prod_coca.rename(columns=dict(zip(prod_coca.columns, variables)))


prod_coca
```

```python
prod_coca['País'] = prod_coca['País'].str.replace('Colombia a', 'Colombia').str.replace('Peru b', 'Peru')

# Mostrar el DataFrame actualizado
prod_coca
```

```python
# Eliminar espacios en la columna 'País'
prod_coca['País'] = prod_coca['País'].str.strip()

# Obtener las columnas de años
columnas_años = prod_coca.columns[1:]

# Utilizar la función melt para agrupar las columnas de años en la variable "años"
prod_melted = prod_coca.melt(id_vars=['País'], value_vars=columnas_años, var_name='años', value_name='valor')

# Imprimir la base de datos con las columnas agrupadas en "años"
print(prod_melted)
```

```python
paises_unicos = prod_melted['País'].unique()
print(paises_unicos)
```

```python
# Filtrar los datos por país
bolivia = prod_melted[prod_melted['País'] == 'Bolivia (Plurinational State of)']
colombia = prod_melted[prod_melted['País'] == 'Colombia']
peru = prod_melted[prod_melted['País'] == 'Peru']

# Crear la figura y los ejes
fig, ax = plt.subplots(figsize=(8, 5))

sns.set(style='ticks')

# Graficar 
sns.lineplot(data=peru, x='años', y='valor', color='red', linestyle='dotted', label='Peru')
sns.lineplot(data=bolivia, x='años', y='valor', color='blue', linestyle='dashdot', label='Bolivia')
sns.lineplot(data=colombia, x='años', y='valor', color='green', linewidth=0.5, label='Colombia')

# Agregar  marcadores
sns.scatterplot(data=bolivia, x='años', y='valor', color='blue', marker='s', label='Bolivia')
sns.scatterplot(data=colombia, x='años', y='valor', color='green', marker='x', label='Colombia')
sns.scatterplot(data=peru, x='años', y='valor', color='red', marker='o', label='Peru')

# Etiquetas de los ejes y título del gráfico
plt.xlabel('Year')
plt.ylabel('Coca Production')
plt.title('Figure 1: Global illicit cultivation of coca bush, 2009-2020 (hectares)', fontsize=13)

# Alinear los valores del eje Y en posición vertical
plt.yticks(rotation='vertical')

# Quitar los top y right spines
sns.despine()

# Ajustar  Y
plt.ylim(0, max(prod_melted['valor']) * 1.1)

# Mejorar la distribución de los elementos del gráfico
plt.tight_layout()

# Colocar la leyenda en la parte inferior
plt.legend(loc='lower center', bbox_to_anchor=(0.5, -0.3), ncol=3)

# Agregar estilo de cuadrícula con líneas horizontales
plt.grid(axis='y', linestyle='--', linewidth=0.5)



#notas y fuente
notes = '''Notes: Different area concepts and their effect on comparability were presented in the 
World Drug Report 2012.'''
source = '''Source: Plurinational State of Bolivia: national illicit crop monitoring system supported 
by the United Nations Office on Drugs and Crime (UNODC)'''

plt.text(0, -0.55, notes, ha='left', va='center', transform=ax.transAxes, fontsize=12, linespacing=1.5)
plt.text(0, -0.4, source, ha='left', va='center', transform=ax.transAxes, fontsize=12, linespacing=1.5)

# Mostrar el gráfico
plt.show()

# save plot

fig.savefig(r'../../output/plots/G9_plot_Productionpy.png', dpi=800, bbox_inches='tight')
```

# Relizar un gráfico con los datos de erradicación de hectareas de hoja de Coca.

```python
columnas_a_leer = "C, F:Q"
err_coca = pd.read_excel("../../data/produccion_coca/6.1.2_-_Eradication_of_coca_bush.xlsx", usecols=columnas_a_leer,  skiprows=lambda x: x not in [0,1, 2, 3, 4], header=1)
err_coca = err_coca.rename(columns={'Unnamed: 2': 'País'})
err_coca

#Omitimos el título, la fuente.
#También omitimos el método de erradicación porque es manual para todos y la unidad esta en hectáreas para todos
```

```python
columnas_años = err_coca.columns[1:]

# Utilizar la función melt para agrupar las columnas de años en la variable "años"
data_melted = err_coca.melt(id_vars=['País'], value_vars=columnas_años, var_name='años', value_name='valor')

# Imprimir la base de datos con las columnas agrupadas en "años"
print(data_melted)
```

```python
# Filtrar los datos por país
bolivia_data = data_melted[data_melted['País'] == 'Bolivia (Plurinational State of)']
colombia_data = data_melted[data_melted['País'] == 'Colombia']
peru_data = data_melted[data_melted['País'] == 'Peru']

# Crear la figura y los ejes
fig, ax = plt.subplots(figsize=(8, 5))

# Configurar el estilo de seaborn
sns.set(style='ticks')

# Graficar las líneas para Bolivia, Colombia y Perú
sns.lineplot(data=peru_data, x='años', y='valor', color='red', linestyle='dotted', label='Peru')
sns.lineplot(data=bolivia_data, x='años', y='valor', color='blue', linestyle='dashdot', label='Bolivia')
sns.lineplot(data=colombia_data, x='años', y='valor', color='green', linewidth=0.5, label='Colombia')


# Agregar puntos para cada año con diferentes marcadores
sns.scatterplot(data=bolivia_data, x='años', y='valor', color='blue', marker='s')
sns.scatterplot(data=colombia_data, x='años', y='valor', color='green', marker='x')
sns.scatterplot(data=peru_data, x='años', y='valor', color='red', marker='o')

# Etiquetas de los ejes y título del gráfico
plt.xlabel('Year')
plt.ylabel('Coca Erradication')
plt.title('Figure 2: Reported eradication of coca bush, 2009-2020', fontsize=13)

# Alinear valores del eje Y en posición vertical
plt.yticks(rotation='vertical')

# Quitar los top y right spines
sns.despine()

# Ajustar los límites del eje Y
plt.ylim(0, max(data_melted['valor']) * 1.1)  # Ajustamos el factor 1.1 según los datos

# Mejorar la distribución de los elementos del gráfico
plt.tight_layout()

# Colocar la leyenda en la parte inferior
plt.legend(loc='lower center', bbox_to_anchor=(0.5, -0.3), ncol=3)

#estilo
# Agregar estilo de cuadrícula con líneas horizontales
plt.grid(axis='y', linestyle='--', linewidth=0.5)

#Agregar notas y uente
#notas y fuente
notes = '''Notes: The totals for Bolivia (Plurinational State of) and Peru include voluntary and forced eradication.'''
source = '''Source: United Nations Office on Drugs and Crime annual report questionnaire and government reports.'''

plt.text(0, -0.45, notes, ha='left', va='center', transform=ax.transAxes, fontsize=12, linespacing=1.5)
plt.text(0, -0.4, source, ha='left', va='center', transform=ax.transAxes, fontsize=12, linespacing=1.5)



# Mostrar el gráfico
plt.show()
# save plot

fig.savefig(r'../../output/plots/G9_plot_Erradicationpy.png', dpi=800, bbox_inches='tight')
```

# REGEX

```python
  import pandas as pd
import re 
```

```python
#Primero corremos la base de datos
#Usamos regex para cambiar el formato de las coordenadas a uno de coordenadas geográficos (i.e -11.25, -69.56). 
#Nos guiamos de la aplicación a las coordenadas del Capac Ñan usada en el script de python.
```

```python
df= pd.read_excel('../../data/metropolitano.xlsx')
df
```

```python
df['sur_latitud'] = df['sur_latitud'].map(lambda x: re.sub('[^0-9°-]','',x)).map(lambda x: re.sub('[°]','.',x))
```

```python
df['oeste_longitud'] = df['oeste_longitud'].map(lambda x: re.sub('[^0-9°-]','',x)).map(lambda x: re.sub('[°]','.',x))
df
```
