```python
# Regex
# 1. 
# La base de datos metropolitano.xlsx posee información de latitud y longitud de las estaciones del metropiltano y dos lineas de alimentadores. Usar regex para cambiar el formato de las coordenadas a uno de coordenadas geográficos (i.e -11.25, -69.56). Guiese de la aplicación a las coordenadas del Capac Ñan usada en el script de python.


import pandas as pd
import numpy as np
import re  # for regular expressions (REGEX)
import os  # for directorio
import unidecode # to drop tildes
from datetime import datetime  # library for time
import warnings
import math
warnings.filterwarnings('ignore') 

# Subimos la base de datos

df = pd.read_excel("/Users/kubo/Documents/GitHub/1ECO35_2023_1/data/metropolitano.xlsx") 


# Definimos el patrón que busca los números de grados, minutos y segundos
pattern = r'(-?\d{1,3})\s*°\s*(\d{1,2})\s*\'\s*(\d{1,2}\.\d{1,2})\"'

# Función para cambiar el formato de las coordenadas
def format_coordinates(match):
    degrees = float(match.group(1)) # agrupmoos los grados
    minutes = float(match.group(2)) # agrupoamos los minutos
    seconds = float(match.group(3)) # agrupamos los segundos
    
# Transformamos a coordenadas decimales    
    decimal_coordinates = degrees + (minutes / 60) + (seconds / 3600)
    
# Acotamos que solo queremos 2 decimales, tal como muestra el ejemplo del enunciado
    return f"{decimal_coordinates:.2f}"

# Usamos re.sub para aplicar los cambios hechos por decimal_coordinates a todas las entradas de ambas columnas 
df['sur_latitud'] = df['sur_latitud'].apply(lambda x: re.sub(pattern, format_coordinates, str(x)))
df['oeste_longitud'] = df['oeste_longitud'].apply(lambda x: re.sub(pattern, format_coordinates, str(x)))


# Juntamos los elementos en un parentesis (sur_latitud , oeste_longitud)
df['coordenadas'] = '(' + df['sur_latitud'].apply(lambda x: re.sub(pattern, format_coordinates, str(x))) + ', ' + df['oeste_longitud'].apply(lambda x: re.sub(pattern, format_coordinates, str(x))) + ')'

print(df['coordenadas'])


```
