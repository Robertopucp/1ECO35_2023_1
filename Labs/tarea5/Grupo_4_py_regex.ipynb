```python
def convert_gps(x):
    signo = -1 if re.search('[swSWoO-]', x) else 1
    regex = re.compile(r'[^\d+\s°\']')
    components = re.split(r'[°\'"]', regex.sub('', x))
    
    if len(components) == 3:
        h, m, s = components
        d = ''
    elif len(components) == 4:
        h, m, s, d = components
    else:
        return None
    
    return signo * (float(h) + float(m) / 60 + float(s) / 3600)
```

```python
data2['latitud'] = data2.sur_latitud.map(convert_gps)
data2['longitud'] = data2.oeste_longitud.map(convert_gps)
```

```python
#Limpiar el nombre de los estudiantes.

def limpiar_nombre(nombre):
    # Eliminar caracteres especiales y números
    nombre = re.sub(r'[^a-zA-ZñÑáéíóúÁÉÍÓÚ\s]', '', nombre)
    
    # Eliminar espacios en exceso y al principio/fin del nombre
    nombre = re.sub(r'\s+', ' ', nombre).strip()
    
    # Separar nombre y apellidos
    partes = nombre.split(' ')
    
    # Eliminar palabras vacías y apellidos con un solo carácter
    partes = [parte for parte in partes if parte and len(parte) > 1]
    
    # Unir nombre y apellidos
    nombre_limpio = ' '.join(partes)
    
    return nombre_limpio

# Aplicar la función limpiar_nombre a la columna 'NAME' en la base de datos 'base2'
base2['nombres_limpios'] = base2['NAME'].map(limpiar_nombre)
```

```python
base2
```

```python
# Función para limpiar la fecha de nacimiento

def clean_birth_date(date):
    pattern = r'\d{2}/\d{2}/\d{4}'  # Patrón de fecha dd/mm/yyyy
    match = re.search(pattern, str(date))
    if match:
        return match.group()
    else:
        return ''

# Convertir la columna a cadenas y limpiar la fecha de nacimiento
base2['BORN_DATE'] = base2['BORN_DATE'].astype(str).apply(clean_birth_date)

```

```python
base2
```

```python
# Similar al item anterior, limpiar la edad

def clean_age(age):
    pattern = r'\d+'  # Patrón de dígitos
    matches = re.findall(pattern, str(age))
    if matches:
        return int(matches[0])
    else:
        return None
```

```python
base2['AGE'] = base2['AGE'].apply(clean_age)
```

```python
## Recuperar el nombre y edad correctos, y reemplazarlo en las variables de nombre (NAME) y edad respectivamente (AGE).

# Función para extraer nombres corregidos
def extract_name_corrected(text):
    if isinstance(text, str):
        match = re.search(r'nombre correcto es ([A-Za-z\s]+)', text)
        if match:
            return match.group(1)
    return np.nan

# Función para extraer edades corregidas
def extract_age_corrected(text):
    if isinstance(text, str):
        match = re.search(r'edad correta es (\d+)', text)
        if match:
            return int(match.group(1))
    return np.nan

# Crear las nuevas columnas 'NAME_CORRECTED' y 'AGE_CORRECTED'
base2['NAME_CORRECTED'] = base2['observaciones'].apply(extract_name_corrected)
base2['AGE_CORRECTED'] = base2['observaciones'].apply(extract_age_corrected)
```

```python
## Reemplazamos solo los nombres correctos

## Obtener una copia de la columna 'NAME'
base2['NAME'] = base2['NAME'].copy()

## Reemplazar los valores no nulos de 'NAME_CORRECTED' en 'NAME'
base2.loc[base2['NAME_CORRECTED'].notnull(), 'NAME'] = base2['NAME_CORRECTED']
```

```python
## Reemplazamos solo las edades correctas

## Obtener una copia de la columna 'AGE'
base2['AGE'] = base2['AGE'].copy()

## Reemplazar los valores no nulos de 'AGE_CORRECTED' en 'AGE'
base2.loc[base2['AGE_CORRECTED'].notnull(), 'AGE'] = base2['AGE_CORRECTED']
```

```python
## Crear una variable con la cantidad de hermana/os del estudiante

# Definir función para extraer la cantidad de hermanas y hermanos
def extract_siblings(text):
    match_sisters = re.search(r'tiene (\d+) hermanas?', str(text))
    match_brothers = re.search(r'tiene (\d+) hermanos?', str(text))
    
    sisters = int(match_sisters.group(1)) if match_sisters else 0
    brothers = int(match_brothers.group(1)) if match_brothers else 0
    
    return sisters, brothers

# Aplicar la función a la columna 'observaciones' y guardar los resultados en nuevas columnas 'SISTERS' y 'BROTHERS'
base2['SISTERS'], base2['BROTHERS'] = zip(*base2['observaciones'].apply(extract_siblings))
```
