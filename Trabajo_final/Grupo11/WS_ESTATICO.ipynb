```python
# dict para almacenar los tables 
#notar que debemos ahcer cambio d efechas y tenemos ahora 21
all_tables = {}

for i in tqdm(range(1319)):
    
    table_html = driver_elecciones.find_element(By.ID, "ember114").get_attribute('innerHTML')
    #extrae todo el elemento dentro de esta division la cual contiene la tabla
    table = pd.read_html(table_html)#detecta la tabla y lo convierte a data frame
    table = table[0] #diccionario vacio
    all_tables[f"table_page_{i}"] = table  # llena el diccionario 
    driver_elecciones.find_element(By.XPATH, "//a[@aria-label='Siguiente']").click() #le doy click en siguiente
    time.sleep(2)

    
#driver_visitas.quit()  # cierra todos los bwrosers ( ventanas )
#driver_visitas.close()  # cierra el browser donde estamos trabajando
```
