```python
#Se procede a cargar las librerías necesarias
import pandas as pd

from selenium import webdriver  # Nos permite navegar en la web
import re  # regrex
import time  # control del tiempo 
import unidecode # tiene muchas utilidades pero lo utilizaremos para retirar tildes 
from selenium.webdriver.support.ui import Select  # selccionar opciones de las lista desplegable
from selenium.webdriver.common.by import By  # permite seleccionar los elementos en un html
from selenium.webdriver.common.keys import Keys  # ingresar datos a un formulario
from selenium.common.exceptions import NoSuchElementException
from tqdm import tqdm # contabilizar tiempo en el loop
from selenium.webdriver.common.action_chains import ActionChains # movilizar la pantalla

# simulador de webdriver 
from webdriver_manager.chrome import ChromeDriverManager # usar diferentes versiones del browser

import warnings
warnings.filterwarnings('ignore') # eliminar warning messages 
```

```python
# se define el webdriver a utilizar, le ponemos nombre "driver_elecciones"

driver_elecciones = webdriver.Chrome("chromedriver.exe")

# El browser del buscador se llama ahora driver_elecciones 

driver_elecciones.maximize_window()

url = 'https://venezuela360.org/arcgis/apps/sites/#/infraestructura-geoespacial/datasets/3d0308d198ba422fbdabd2e1859eb577/data'

driver_elecciones.get( url )
```

```python
# creación de nuevo XPATH para cambiar de tabla en tabla, a partir del siguiente elemento
# <a href="" aria-label="Siguiente" data-ember-action="" data-ember-action-109196="109196">›</a>
```

```python
#XPATH --> "//a[@aria-label='Siguiente']"
```

```python
#se verifica que con este codigo se de CLICK a la siguiente página
driver_elecciones.find_element(By.XPATH, "//a[@aria-label='Siguiente']").click()
```

```python
all_tables = {}
num_tables = 1320

for i in tqdm(range(num_tables)):
    
    table_html = driver_elecciones.find_element(By.ID, "ember114").get_attribute('innerHTML')
    #extrae todo el elemento dentro de esta division la cual contiene la tabla
    table = pd.read_html(table_html)#detecta la tabla y lo convierte a data frame
    table = table[0] #diccionario vacio
    all_tables[f"table_page_{i}"] = table
    
    try: 
        if i < len(range(num_tables - 1)):
            driver_elecciones.find_element(By.XPATH, "//a[@aria-label='Siguiente']").click() #le doy click en siguiente
            time.sleep(2)
        else:
            pass
    
    except:
        next_button_position = 6 + i - 1315  # Ajuste de posición a partir de la tabla 1315
        button_xpath = f"(//ul[@class='pagination']/li[not(contains(@class, 'active'))])[{next_button_position}]/a"
        driver_elecciones.find_element(By.XPATH, button_xpath).click()      
    
 
```
