```python
# Primera regrecion 

A = df['took_year']
B = sm.add_constant(df.loc[:,["treat2016", "yr_2016", "treatment_class"]])
```

```python
print(ols_model.summary())
```

```python
ols_model_rb = sm.OLS(A, A).fit(cov_type='cluster', cov_kwds={'groups': df['class_fe2']})
print(ols_model_rb.summary())
```
