```python
driver_venezuela = webdriver.Chrome("chromedriver.exe")


```

```python
time.sleep (2)

# El bwoser del buscador se llama drive-visitas

driver_venezuela.maximize_window()

link = 'https://venezuela360.org/arcgis/apps/sites/#/infraestructura-geoespacial/datasets/3d0308d198ba422fbdabd2e1859eb577/data'

driver_venezuela.get(link)
```

```python
# cantidad de páginas 
pagination_element = driver_venezuela.find_elements(By.CLASS_NAME, "pagination")[-1]
page_numbers = re.findall(r'\d+', pagination_element.text)
total_pages = int(page_numbers[-1])
print("Número total de páginas:", total_pages)

```

```python
# click en siguientes páginas
driver_venezuela.find_element(By.XPATH, "//a[contains(@aria-label, 'Siguiente')]").click()
time.sleep(5)
```

```python
# dict para almacenar los tables 

all_tables = {}

for i in tqdm(range(1318)):
    table_html = driver_venezuela.find_element(By.CLASS_NAME, "table-responsive").get_attribute('innerHTML')
    table = pd.read_html(table_html)
    table = table[0]
    all_tables[f"table_page_{i+2}"] = table  # llena el diccionario 
    driver_venezuela.find_element(By.XPATH, "//a[contains(@aria-label, 'Siguiente')]").click()
    time.sleep(0.1)
    
   

```

```python
all_tables.keys()
```

```python
total_visitas_data = pd.concat( all_tables.values() ).reset_index( drop = True )
total_visitas_data
```
