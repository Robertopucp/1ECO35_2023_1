## Reshape 

### Panel de ENAHO correspondiente al módulo 84 (Sumaria)

- Seguimiento al msimo hogar en difenrentes puntos del tiempo
- Hogares monitoreados en 3 años conseciutivos 2016-2020

- Base de datos forma WIDE

```python
panel = pd.read_stata("../../data/panel_2016_2018.dta",
                      convert_categoricals=False)

# Aqui amos a encontrar varaibles para cada año. Por ejemplo pobreza_16, pobreza_17, pobreza_18,..., pobreza_20

# identificador del hogar: Conglomerado, vivienda y hogar 
```

```python
panel
```

```python
# nombre de las variables en minuscula

panel.columns = map(str.lower, panel.columns)  


print( panel.columns)
```

```python
# Objetivo es quedarme solo con algunas variables 

# Se filtra variables selecionadas 

index_columns = np.where( panel.columns.str.contains(
    '(año)|(^conglome)|(vivienda)|(hogar)|(estrato_)|(mieperho)|(gashog2d)|(pobreza_)|(factor07)',
    regex=True))[0]

# [0] extraer la lista de posiciones dentro de la tupla con las psicicones de las columnas que satisfacen la condición 
```

```python
index_columns
```

```python

# index_columns: se guarda las posiciones de columnas de aquellas variables que satisface 
# con algunos de los casos del regex

panel = panel.iloc[:,index_columns] 


# Rename las  variables de año 
    
panel.rename(columns = {'año_16':'year_16', 'año_17':'year_17', 'año_18':'year_18', 'año_19':'year_19',
                       'año_20':'year_20', 'conglome':'cong', 'vivienda':'viv'}, inplace = True)
```

```python
panel
```

```python

# Se crea un identificados del hogar 

panel['hog'] = np.arange(1, panel.shape[0] + 1)
```

```python
panel
```

```python
# reordenando las columnas 

cols = panel.columns.tolist() # nombre de variable a lista 
```

```python
cols[0:2]
```

```python
[ cols[-1] ]
```

```python
cols[2:-1]
```

```python
cols[0:2]+ [ cols[-1] ] + cols[2:-1]
```

```python

new_cols_orders = cols[ 0:2 ] + [cols[ -1 ]] + cols[ 2:-1 ] 

# cols[ 0:2 ] dos primeras columnas (conglome, vivienda)
# [cols[ -1 ]] ultima columna (hog)
# cols[ 2:-1 ] desde la tercera hasta la penultima
new_cols_orders 
```

```python
panel = panel.loc[:, new_cols_orders]

# uso .loc por que usaré el nombre de las columnas 

panel
```

```python
## Reshape Wide to long 

# reshape_panel = pd.wide_to_long(panel, stubnames = new_list,
#                                 i = ['cong', 'viv', 'hog'] , j = 'period' , sep = '_').reset_index()


# new_list: debe contener las variables en comun en todos los años (sin _año)

panel.columns[3::]
```

```python
filter_list = list( map( lambda x: re.sub('.{3}$',"", x) , list(panel.columns)[3::] )   )
    
#se reemplaza los tres ultimos caracteres 
#(como se usa el carcater especial .), se toma cualquier tipo de caracter ya sea espacio, numero, letras, #!", etc

# Se puede ibservar que filter list tiene duplicados

# Se procede a elimitar los duplicados en la lista

new_list = list(dict.fromkeys(filter_list))
```

```python
new_list
```

```python
## Reshape Wide to long 

reshape_panel = pd.wide_to_long(panel, stubnames = new_list, i = ['cong', 'viv', 'hog'] , 
                                j = 'period' , sep = '_').reset_index()


# pd.wide_long( base de datos, lista de variables, lista de identificadores, periodo, sep)
```

```python
reshape_panel
```

```python
# drop a las siguientes variables

del reshape_panel['cong']
del reshape_panel['viv']
del reshape_panel['hog']
del reshape_panel['period']
```

## Uso de pivot y melt

```python
bdata = pd.DataFrame({'geo_id': ['01', '01', '02', '02', '04', '04', '05', '05', '06', '06'],
                      'name': ['Alabama', 'Alabama', 'Alaska', 'Alaska', 'Arizona', 'Arizona', 'Arkansas', 'Arkansas', 'California', 'California'],
                      'variable': ['income', 'rent', 'income', 'rent', 'income', 'rent', 'income', 'rent', 'income', 'rent'],
                      'estimate': [24476, 747, 32940, 1200, 27517, 972, 23789, 709, 29454, 1358],
                      'moe': [136, 3, 508, 13, 148, 4, 165, 5, 109, 3]})
```

```python
bdata

# income: annual income
# rent monthly rent
# moe: standar error
```

```python
# Pivot_table equivalente a pivot_wider

# Tantas columnas según cantidad de categorías

df1 = bdata.pivot_table(index= ['geo_id','name'], columns='variable', values= ['estimate', 'moe'])
```

```python
df1
```

```python
df1.columns.values
```

```python
# rename columns dataframe

df1.columns = ['_'.join(col).strip() for col in df1.columns.values]  # list comprenhension 
df1 = df1.reset_index()

# '_'.join(col).strip()  unir elementos de la tupla usando el guion bajo
```

```python
df1
```

```python
# Melt es el aquivalente de pivot_longer de r
# Es necesario un paso adicional. Aplicar Pivot pues tenemos varias columnas de values (estiamte y moe)

## Tantas filas por unidad de observación según categorias

new_df = pd.melt(df1, id_vars=['geo_id','name'], # identificadores 
                 value_vars=['estimate_income', 'estimate_rent','moe_income','moe_rent'], # values varis
                 var_name='type', value_name='value' , col_level=None, ignore_index=True)

new_df
```

```python
new_df[['est','category']] = new_df['type'].str.split('_', expand=True)
del new_df['type']

new_df
```

```python
new_df = new_df.pivot(index = ['geo_id','name','category'], columns = 'est', values = 'value').reset_index()
```
