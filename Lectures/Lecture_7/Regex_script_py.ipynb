# re.search

- Busqueda de patrones para extraer de un texto 

```python
# extracción del nombre de distrito y región 


x = "AVENIDA LA PAZ CUADRA 3 LA PERLA CALLAO CALLAO distrito LA PERLA Region El CALLAO"

print(re.search('[D/d]istrito ([\w+\-\s]+) [R/r]egion ([\w+\s]+)', x).group(1))

re.search('[D/d]istrito ([\w+\-\s]+) [R/r]egion ([\w+\s]+)', x).group(2)
```

```python
# () permite capturar lo que me interesa. Puede acceder a cada grupo de interés mediante .group()
```

```python

def dist_region(x):
    regex_area = re.compile('[D/d]istrito ([\w+\-\s]*) [R/r]egion ([\w+\s]*)')
    output =  regex_area.search(x)
    
    return output.group(1), output.group(2)
 
     
data['distrito'] = data['dirección'].apply(lambda x: dist_region(x)[0])
data['distrito']
```

```python
data['region'] = data['dirección'].apply(lambda x: dist_region(x)[1])
data['region'] 
```

```python
#Extracción del número telefónico

data['telefono']
```

```python
def phone(x):
    
    try:
        match_regex = re.compile('(\d+\-\d+)')
        match = match_regex.search(str(x))
    
        return match.group(1)
    
    # Si no encuentra un grupo por precencia de missing, se ejecuta la siguiente linea
    
    except:
        
        None
```

```python
data['phone'] = data['telefono'].apply(lambda x: phone(x))
data['phone']
```

```python
def reso_info(x):
    
    regex_search = re.compile('([0-9]+)-([0-9]+)\s([A-Z]+)$')
    match = regex_search.search(x)
    
    return match.group(1), match.group(2), match.group(3)
```

```python
# Código de la resolución

data['code_res'] = data['resolucion'].apply(lambda x: reso_info(x)[0])
data['code_res']
```

```python
# Año de emisión de la resolución

data['year_res'] = data['resolucion'].apply(lambda x: reso_info(x)[1])
data['year_res']
```

```python
# Entidad pública que emite la resolución

data['entidad_res'] = data['resolucion'].apply(lambda x: reso_info(x)[2])
data['entidad_res']
```

# Look around

- Extrae lo que extá alrededor de un patrón de texto

```python
# positive lookahead (?=)

regex = re.compile("\d+\:\d+(?= am)")

data['apertura1'] = data['horario'].apply(lambda x: re.search("\d+\:\d+(?= am)",x).group())
data['apertura1'] 
```

```python
# positive lookbehind (?<=)

regex = re.compile("(?<=apertura )[\d+\:]+")
data['apertura3'] = data['horario'].apply(lambda x: regex.search(x).group())
data['apertura3']
```

```python
# negative lookbehind (?<!)

regex = re.compile("(?<=apertura )[\d+\:]+")
data['cierre1'] = data['horario'].apply(lambda x: regex.search(x).group())
data['cierre1']
```

```python
# negative lookbahead (?!)

regex = re.compile("[\d*\,]+(?!\$)")
data['pres_soles'] = data['presupuesto'].apply(lambda x: regex.search(x).group())
data['pres_soles']
```

```python
# retirar tildes del texto

data['presupuesto'] = data['presupuesto'].apply(unidecode.unidecode)
```

```python
def convert_gps(x):
        
    signo = -1 if re.search('[swSWoO-]', x ) else 1
    
    regex = re.compile('[^\d+\s\.]') # Todo lo que sea diferente a digitos, especio y el punto
    h, m , s = str.split( regex.sub('', x ) )
    
    return signo*(float(h) + float(m)/60 + float(s)/3600)
    
```

```python
data2['latitud'] = data2.Coordinates.map(convert_gps)
data2['longitud'] = data2.Coordinates2.map(convert_gps)
```
