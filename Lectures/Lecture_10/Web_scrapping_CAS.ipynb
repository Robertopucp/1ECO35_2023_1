```python
driver = webdriver.Chrome("chromedriver.exe")
```

```python
# URl de la p√°gina web

url = 'https://www.mef.gob.pe/contenidos/cas/procesos_2016_concluidas.php'

driver.get( url )  # la pagina web se dirige al url 

time.sleep(2)
```

```python
# Lista de los objetos que poseen un link de descarga .pdf

lista_links = driver.find_elements(By.XPATH, "//a[@class='Estilo4']")
```

```python
lista_links
```

```python
# se extrae el texto asignado a los links 

lista_links[4].text
```

```python
# Se puede extraer el link de descarga de cada elemento html usando .get_attribute("href")

lista_links[0].get_attribute("href")
```

```python
# Filter the links to PDF files
pdf_links = [link.get_attribute("href") for link in lista_links if link.get_attribute("href").endswith(".pdf")]

# .endswith(".pdf") el link culmina en .pdf
```

```python
pdf_links
```

```python
# Iterate over the PDF links and download each file
for pdf_link in pdf_links:
    
    response = requests.get(pdf_link) # request cada link 
    
    with open(pdf_link.split("/")[-1], "wb") as f:  # se crea el nombre del pdf y se guarda el archivo
        f.write(response.content)
```
