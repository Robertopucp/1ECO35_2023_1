```python
# Configuraci칩n de la descarga

chrome_options = Options()
chrome_options.add_experimental_option('prefs', {
"download.default_directory": "C:/Users", # COLOCAR EL EQUIVALENTE EL APPLE
"download.prompt_for_download": False, # descarga pdf autom치tico
"download.directory_upgrade": True,
 "plugins.always_open_pdf_externally": True # no abre una ventana del pdf
})
```

```python
# Similar de chrome y se coloca la configuraci칩n

service=Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options = chrome_options)
```

```python
driver.maximize_window()
```

```python
# URl de la p치gina web

url = 'https://www.mef.gob.pe/contenidos/cas/procesos_2016_concluidas.php'

driver.get( url )  # la pagina web se dirige al url 

time.sleep(2)
```

```python
# Lista de los objetos que poseen un link de descarga .pdf

lista_links = driver.find_elements(By.XPATH, "//a[@class='Estilo4']")
```
